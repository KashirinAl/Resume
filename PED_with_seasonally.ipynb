```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_sales = pd.read_excel('Данные для кейса_версия 2021.xlsx', sheet_name='Итоговый сет' )
df_sales['AIV'] = df_sales['Оборот без НДС'] * 1.2 / df_sales['Количество']
```

```python
df_sales_1 = df_sales.groupby(['Тов.иер.ур.4','Товар','Месяц'])[['Количество','Оборот без НДС','Маржа']].sum().reset_index()

df_sales_1['AIV'] = df_sales_1['Оборот без НДС'] * 1.2 / df_sales_1['Количество']
df_sales_1['AIM'] = df_sales_1['Маржа'] / df_sales_1['Количество']

df_sales_1[['Тов.иер.ур.4', 'Товар', 'Месяц', 'Количество', 'AIV', 'AIM']].to_excel('Продажи_товаров.xlsx')
df_sales_1
```

```python
month_sales = df_sales.groupby(['Месяц','Филиал'])[['Количество']].sum()
month_sales  = month_sales.reset_index().pivot_table(
    index='Месяц',
    columns='Филиал',
    values='Количество'
)

month_sales_vertical = pd.DataFrame( [], columns=['Месяц','Сезонность','Филиал'] )
for colum in month_sales.columns:
    month_sales[colum] =  month_sales[colum] / month_sales[colum].mean()
    
    pd_f = pd.DataFrame( month_sales[colum] ).reset_index()
    pd_f.columns = ['Месяц','Сезонность']
    pd_f['Филиал'] = colum
    
    month_sales_vertical = pd.concat([month_sales_vertical, pd_f])
    
```

```python
month_sales.to_excel('sales_by_season.xlsx')
```

```python
df_sales_season = df_sales.merge( month_sales_vertical,
              left_on=['Месяц','Филиал'], right_on=['Месяц','Филиал'],
               suffixes=('','_в_месяц')
          )

df_sales_season['Количество_сез'] = df_sales_season['Количество'] / df_sales_season['Сезонность']

df_sales_season
```

```python
df_sales_season_1 = df_sales_season[['Филиал','Тов.иер.ур.4','Товар','Месяц','AIV','Количество_сез']].sort_values(by=['Товар','Месяц'])
df_sales_season_1
```
